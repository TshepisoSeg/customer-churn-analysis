## Sales Data Analysis
Analyzing monthly product sales in 3 different ways. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import CSV of monthly sales
df = pd.read_csv("MonthlyProductSales.csv", encoding = "ISO-8859-1")
```

```python
df.groupby(df['Month of Order Date']).sum(numeric_only=True).plot()
```

```python
df.groupby(df['Month of Order Date']).sum(numeric_only=True).plot(kind='hist', bins=10)
```

```python
df.groupby(df['Product Name']).sum(numeric_only=True).sort_values(by='Sales').tail(10).plot(kind='barh')
```
